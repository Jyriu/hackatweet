```python
import requests
import pandas as pd
import random
import json
from faker import Faker
from pymongo import MongoClient

```

```python
# Connexion à MongoDB
MONGODB_URI = "mongodb+srv://samiyezza:8a08ad81@cluster0.uyxehij.mongodb.net/hackatweet"
client = MongoClient(MONGODB_URI)
db = client.get_database("hackatweet")

# Configuration de l'API
BASE_URL = "http://localhost:5001/api"
REGISTER_URL = f"{BASE_URL}/auth/register"
TWEET_URL = f"{BASE_URL}/tweet/createTweet"

```

```python
# Génération de faux utilisateurs
fake = Faker()
nombre_utilisateurs = 30
utilisateurs = []

def creer_utilisateur():
    email = fake.email()
    nom = fake.last_name()
    prenom = fake.first_name()
    username = fake.user_name()
    password = "password123"
    
    return {
        "email": email,
        "password": password,
        "nom": nom,
        "prenom": prenom,
        "username": username
    }

```

```python
# Création des utilisateurs
for _ in range(nombre_utilisateurs):
    user_data = creer_utilisateur()
    response = requests.post(REGISTER_URL, json=user_data)
    if response.status_code == 201:
        user_info = response.json()
        utilisateurs.append({
            "id": user_info["user"]["id"],
            "token": user_info["token"]
        })
        print(f"Utilisateur {user_data['username']} créé avec succès.")
    else:
        print(f"Erreur lors de la création de {user_data['username']} {response.status_code}.")

```

```python
# Chargement du dataset de tweets
dataset_path = "./datas/dataset.csv"
df = pd.read_csv(dataset_path, encoding="latin1")

# Nettoyage et sélection des tweets
df = df[['tweet_text']].dropna()
tweets = df['tweet_text'].tolist()

```

```python
# Envoi des tweets à l'API
for tweet in tweets:
    user = random.choice(utilisateurs)  # Choisir un utilisateur aléatoire
    headers = {"Authorization": f"Bearer {user['token']}"}
    data = {"text": tweet}
    response = requests.post(TWEET_URL, json=data, headers=headers)
    if response.status_code == 201:
        print(f"Tweet ajouté par {user['id']}")
    else:
        print(f"Erreur lors de l'ajout du tweet.")
```
